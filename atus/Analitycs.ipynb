```python
import pandas as pd

df = pd.read_csv("../input/atus_anual_2022.csv")
print(df.head())
```

```python
#1-Cargue en un Dataframe los datos de los accidentes de los últimos 3 años
d2020 = pd.read_csv("../input/atus_anual_2020.csv", index_col = False)
d2021 = pd.read_csv("../input/atus_anual_2021.csv", index_col = False)
d2022 = pd.read_csv("../input/atus_anual_2022.csv", index_col = False)

dfult3 = [d2020, d2021, d2022]
dfult3 = pd.concat(dfult3)

print (dfult3)
```

```python
#2-Agregue una columna tipo date que englobe las columnas correspondientes a Año, Mes, Dia, Hora y Minuto.
dfult3['FECHA'] = dfult3.apply(lambda row: f"{row['ANIO']}-{row['MES']}-{row['ID_DIA']} {row['ID_HORA']}:{row['ID_MINUTO']}", axis=1)

dfult3['FECHA'] = pd.to_datetime(dfult3['FECHA'], format='%Y-%m-%d %H:%M')

print(dfult3.head())
```

```python
#3-Agregue una columna para Área cuyos valores serán Urbana o Suburbana dependiendo del área donde ocurrió el accidente.
def detArea(row):
    if row['SUBURBANA'] != 'Sin accidente en esta zona':
        return 'SUBURBANA'
    else:
        return 'URBANA'


dfult3['AREA']= dfult3.apply(detArea,axis=1)
print(dfult3.head())
```

```python
#4-Agregue una columna Zona cuyos valores serán la zona donde ocurrió el accidente
def detZona(row):
    if row['AREA']=='URBANA':
        return row['URBANA']
    else:
         return row['SUBURBANA']

dfult3['ZONA']= dfult3.apply(detZona,axis=1)
print(dfult3.head())
```

```python
#5-Elimine las columnas URBANA y SUBURBANA
dfult3.drop(columns=['URBANA', 'SUBURBANA'], inplace=True)
print(df.head())
```

```python
#6-Determine cual municipio de cada estado tuvo más meses sin accidentes
dfult3['AccidentesPorMes'] = dfult3[['AUTOMOVIL', 'CAMPASAJ', 'MICROBUS', 'PASCAMION', 'OMNIBUS', 'TRANVIA',
                                       'CAMIONETA', 'CAMION', 'TRACTOR', 'FERROCARRI', 'MOTOCICLET', 'BICICLETA',
                                       'OTROVEHIC']].sum() 
Acci=dfult3.groupby(['ID_ENTIDAD','ID_MUNICIPIO','ANIO'])['MES'].value_counts().reset_index() 
AcciMes = Acci.loc[Acci.groupby('ID_ENTIDAD')['count'].idxmin()]
AcciMes.sort_index() 
AcciMes
```

```python
#7-Resuma los accidentes del último año por tipo y que porcentaje representan
TipAcci = d2022['TIPACCID'].value_counts().reset_index()
TipAcci.columns = ['Tipo de Accidente', 'Count']

TotAcci = TipAcci['Count'].sum()
TipAcci['Porcentaje sobre el Total'] = (TipAcci['Count'] / TotAcci) * 100
print(TipAcci)
```

```python
#8-Obtenga la cantidad de muertos y lesionados de los últimos 3 años
totmuer = dfult3['PEATMUERTO'].sum()+dfult3['CICLMUERTO'].sum()+dfult3['OTROMUERTO'].sum()+dfult3['NEMUERTO'].sum()
totles =dfult3['PEATHERIDO'].sum()+dfult3['CICLHERIDO'].sum()+dfult3['OTROHERIDO'].sum()+dfult3['NEHERIDO'].sum()

print("Muertos ultimos 3 años:", totmuer)
print("Heridos ultimos 3 años:", totles)
```

```python
#9 ¿Cómo se compara el número de accidentes de cada año contra el anterior? ¿En qué porcentaje varió?
dfult3['FECHA'] = pd.to_datetime(dfult3['FECHA'])

dfult3['AÑO'] = dfult3['FECHA'].dt.year


TotAcciAño = dfult3['AÑO'].value_counts().reset_index()
TotAcciAño.columns = ['AÑO', 'Total_Accidentes']

TotAcciAño['Porcentaje'] = (TotAcciAño['Total_Accidentes'] / TotAcciAño['Total_Accidentes'].sum()) * 100

dfult3 = dfult3.drop('AÑO', axis = 1)
print(TotAcciAño)
```
